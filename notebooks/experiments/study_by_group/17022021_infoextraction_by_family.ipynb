 ## 1. Information extraction and data tidying. <a name="id1"></a>

```python
import pandas as pd
import openpyxl
import xlrd
import numpy as np
import glob
import matplotlib.pyplot as plt
```

Here we consider to take with the function ``unique()`` all the family products in that year, to make it easier.

```python
data = pd.read_excel('C:/Users/Javier/Documents/GIT/Repositories/TFM-Textil-Santanderina/data/raw/ejemplo_datos.xlsx', engine='openpyxl')
```

```python
list(data.GrupoArticulo.unique())
```

```python
tablas20_ = glob.glob("C:/Users/Javier/Documents/GIT/Repositories/TFM-Textil-Santanderina/data/raw/2016*.xlsx")
families_per_month = np.empty(0)

for f in tablas20_:
    df = pd.read_excel(f, engine='openpyxl')
    df_f = df.loc[df['TipoDoc'].isin(['F'])]
    list_families = list(df_f.GrupoArticulo.unique())
    families_per_month = np.append(families_per_month,list_families)
    print("esto funciona")
```

```python
 families_per_month = families_per_month.astype('int')
```

```python
families_per_month
```

```python
(unicos, cuenta) = np.unique(families_per_month, return_counts=True)
```

```python
unicos
```

```python
cuenta
```

Then here we have how frequently they appear in each month during the business year 2016.

```python
plt.bar(x= unicos, height= cuenta)
```

```python
pmf_group1 = []
pmf_group2 = []
pmf_group3 = []
pmf_group4 = []
pmf_group5 = []
pmf_group9 = []
pmf_group10 = []
pmf_group11 = []
pmf_group12 = []
empty_array = np.empty(0)


for f in tablas20_:
    df = pd.read_excel(f, engine='openpyxl')
    df_f = df.loc[df['TipoDoc'].isin(['F'])]
    for i in unicos:
        fm = empty_array
        df_g = df_f.loc[df_f['GrupoArticulo'].isin([i])]
        fm = np.append(fm,df_g['Importe'])
        eval('pmf_group' + str(i)).append(fm.sum())
```
